# Find and Find All

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/page/forms/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find_all('div', class_ = '')
```

```python
soup.find_all('a')
```
