```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\aryan\OneDrive\Desktop\Data Analyst\python\Python Pandas\Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
#terdapat 2 cara untuk melakukan perhitungan
```

```python
df.groupby('Base Flavor').count()
```

```python
group_by_frame.count()
#untuk melakukan perhitungan maka tinggal ubah saja bagian .count() sesuai yang dibutuhkan seperti mean, max, min, dll
```

```python
df.groupby('Base Flavor').sum()
```
